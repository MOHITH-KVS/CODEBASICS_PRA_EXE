Now using standard deviation using 4


```python
bhps.describe()
```

```python
max = bhps.price.mean() +  bhps.price.std() * 4
```

```python
min = bhps.price.mean() - bhps.price.std() * 4
```

```python
bhps[(bhps.price > max) | (bhps.price < min)]
```

```python
bhps1 = bhps[(bhps.price < max) & (bhps.price > min)]
```
